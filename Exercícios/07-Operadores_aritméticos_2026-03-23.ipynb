## 02 – Obtenha os resultados e retorne o código do objeto em Python dos seguintes cálculos:
a) [2.0, 3.0, 5.0] + [10, 20, 30]

b) lista_a * 2

c) lista_a + lista_b

d) lista_a / (lista_b * lista_c)

```python
lista_a = [2.0, 3.0, 5.0] + [10, 20, 30]
lista_b = lista_a*2
lista_c = lista_a + lista_b
print(lista_a)
print(lista_b)
print(lista_c)
```

```python
lista_d = lista_a/(lista_b*lista_c)
```

As três primeiras operações deram certo uma vez que a soma e multiplicação (esta última, desde que feita utilizando números inteiros) em listas resulta na concatenação desta classe de objetos. A soma adiciona os elementos em sequência, enquanto a multiplicação repete os valores contidos na lista. Contudo, a lista_d não é possível por conta das características de lista ao utilizar operadores aritméticos, como demonstrado a seguir:

```python
lista_demonstracao_erro_1 = [2.0, 3.0, 5.0, 10, 20, 30, 2.0, 3.0, 5.0, 10, 20, 30]*[2.0, 3.0, 5.0, 10, 20, 30, 2.0, 3.0, 5.0, 10, 20, 30, 2.0, 3.0, 5.0, 10, 20, 30]
```

Como mostrado no código acima (que nada mais seria que lista_b*lista_c), a multiplicação só pode ser feita com números inteiros, uma vez que em lista não é a multiplicação de cada valor contido internamente, mas sim a repetição dos elementos.

```python
lista_demonstracao_erro_2 = lista_a*2.0
```

De mesmo modo, a substituição de um número inteiro por um float, mesmo que numericamente sejam iguais (exemplo: 2 e 2.0), também resulta em um erro.

```python
lista_demonstracao_erro_3 = lista_a/lista_b
```

```python
lista_demonstracao_erro_4 = lista_a/2
```

Por fim, pela mesma razão que não é a multiplicação dos valores e sim a repetição concatenando os valores, também não é possível fazer a divisão de listas. Neste caso, mesmo que utilize valores inteiros não funcionará o código, uma vez que a operação de divisão não é suportada para a classe.

## 07 – Remodele (reshape) o array do exercício anterior, para que contenha 4 colunas. Converta este objeto da classe array para DataFrame. Arredonde todos os valores para até 3 casas decimais. Recupere a DataFrame.

```python
# Remodelando o array para ter 4 colunas
lista_array = lista_array.reshape(-1,4)
lista_array
# Convertendo array para DataFrame
df = pd.DataFrame(lista_array)
# Arredondando valores para até 3 casas decimais
df = round(df, 3)

df
```

```python
# ou, alternativamente, convertendo array para DataFrame e arredondando valores direto
df = round(pd.DataFrame(lista_array), 3)
df
```

## 08 – A última instancia desta DataFrame deve conter as médias dos valores contidos por coluna, o que foi determinado nos exercícios anteriores. Instancie um objeto contendo as médias dos valores em cada linha. Insira este objeto como uma nova coluna na DataFrame. Recupere a DataFrame.

```python
df = round(pd.DataFrame(lista_array), 3)

# Média das linhas
media_linhas = df.mean(axis=1)
media_linhas = round(media_linhas, 3)
df = pd.concat([df, media_linhas], axis = 1, ignore_index=True)
df
```
