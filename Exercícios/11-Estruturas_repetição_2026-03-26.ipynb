## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 03 – Escreva o programa em Python:

```python
# Criando lista com 5000 inteiros
lista_5000 = []
x = 0
while x<5000:
    lista_5000.append(x)
    x = x+1

# Contando pares e ímpares
par = 0
impar = 0

for i in lista_5000:
    if i % 2 == 0:
        par += 1
    else:
        impar += 1

print(par)
print(impar)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
lista_5000 = []
for x in range(5000):             # Trocando while por for
    lista_5000.append(x)
    x = x+1


par = 0
impar = 0
i = 0

while i < len(lista_5000):        # Trocando for por while
    if lista_5000[i] % 2 == 0:
        par += 1
        i = i+1
    else:
        impar += 1
        i = i+1

print(par)
print(impar)
```

```python
df = pd.read_csv('df_genes.csv', index_col='Unnamed: 0')
df
```

## 07 – Escreva o programa em Python:

```python
x = 0
quais_pares = []
quantos_impares = 0

while x < len(df['Número de Caracteres']):
    if df.iloc[x, 0] % 2 == 0:
        quais_pares.append(int(df.iloc[x, 0]))
        x = x+1
    else:
        quantos_impares = quantos_impares + 1
        x = x+1

print("Quais são pares =", quais_pares)
print("Quantos são ímpares =", quantos_impares)
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
quais_pares = []
quantos_impares = 0

for x in range(len(df['Número de Caracteres'])):
    if df.iloc[x, 0] % 2 == 0:
        quais_pares.append(int(df.iloc[x, 0]))
    else:
        quantos_impares = quantos_impares + 1

print("Quais são pares =", quais_pares)
print("Quantos são ímpares =", quantos_impares)
```
