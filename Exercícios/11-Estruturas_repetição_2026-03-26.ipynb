## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Primeiro deve gerar um objeto vazio "z" da classe lista, que servirá para adicionar os valores inteiros. Também deve instanciar um contador "x" equivalente ao inteiro 0. <br>
"while" permitirá adicionar os valores nesta lista indo de 0 a 4999 (5000 valores), gerando uma iteração que permite a adição sequencial dos números inteiros (em que o próximo número seja igual ao anterior + 1). Determinará que a adição continuará ocorrendo até que o valor do contador x, estabelecido previamente, seja igual a 5000. <br>
Dentro de "while", ele primeiro deve adicionar à lista z o valor do contador (.append(x)). Com isso, a primeira adição será 0. Para que este loop não permaneça infinitamente adicionando 0 a lista (uma vez que x em nenhum momento passará a ser igual a 5000), deve adicionar em seguida a relação que fará o acréscimo sequencial de 1 ao contador x (x = x+1). A cada loop será adicionado um novo valor a lista z (.append(x)), progredindo o contador. Dessa maneira, obterá valores de 0 a 4999 (lista contendo 5000 valores inteiros). <br>
Para contar quantos são pares e quantos são ímpares também se pode utilizar de iteração. Neste caso, para avaliar valor por valor da lista, se cada um destes é par ou ímpar. Uma possibilidade é "for" (para gerar a iteração) para cada valor existente na lista z (usando "in"). Assim analisará todos os valores até não ter mais nenhum restante. Como quer verificar se ele é par ou ímpar, torna-se útil criar uma condicional ("if" e "else"). Antes de estabelecer a condicional e loop, é necessário criar dois objetos equivalentes a 0: um para somar e contar quantos valores são pares (objeto par), e outro quantos são ímpares (objeto ímpar). Dentro do loop, a primeira condicional determinada por "if" pode ser para avaliar o número de pares. Uma forma para descobrir se o número é par é utilizar o operador aritmético %, que retornará o resto de uma divisão. Se dividir por 2 um número par, o resto desta divisão será 0. Logo, basta adicionar o operador relacional de igualdade (==) para verificar se ele é igual a 0. Se for verdadeiro, ele deve adicionar +1 ao valor do objeto par. Portanto, estará contando 1 objeto par, e como está dentro de um loop; sucessivamente a cada vez que essa relação for verdadeira. Caso seja falso, seguirá para o else, que adicionará +1 ao valor do objeto ímpar. Sob mesma lógica, contará quantos são ímpares até não restar nenhum valor para analisar. <br>
Por fim, basta retornar os objetos par e ímpar para obter quantos são pares e ímpares, respectivamente.

## 03 – Escreva o programa em Python:

```python
# Criando lista com 5000 inteiros
lista_5000 = []
x = 0
while x<5000:
    lista_5000.append(x)
    x = x+1

# Contando pares e ímpares
par = 0
impar = 0

for i in lista_5000:
    if i % 2 == 0:
        par += 1
    else:
        impar += 1

print("Pares =", par, "\n", "Ímpares =", impar)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
lista_5000 = []
for x in range(5000):             # Trocando while por for
    lista_5000.append(x)

par = 0
impar = 0
i = 0

while i < len(lista_5000):        # Trocando for por while
    if lista_5000[i] % 2 == 0:
        par += 1
    else:
        impar += 1
    i = i+1

print("Pares =", par, "\n", "Ímpares =", impar)
```

## 07 – Escreva o programa em Python:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df_genes.csv', index_col='Unnamed: 0')
df
```

```python
x = 0
quais_pares = []
quantos_impares = 0

while x < len(df['Número de Caracteres']):
    if df.iloc[x, 0] % 2 == 0:
        quais_pares.append(int(df.iloc[x, 0]))
    else:
        quantos_impares = quantos_impares + 1
    x = x+1

print("Quais são pares =", quais_pares, "\n", "Quantos são ímpares =", quantos_impares)
```
