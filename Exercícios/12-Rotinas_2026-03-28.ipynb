## 01 – Cite três vantagens do uso das funções e duas características que toda função deva ter?

R.: \
A primeira vantagem é a de processamento. Deve-se por não ter que, toda vez que rodar o código, recorrer linha por linha (como em construções com "if" e "while). Será processada uma única linha contendo a função, mesmo que para esta ser estabelecida inicialmente tenha-se usado mais de uma. \
A segunda vantagem é a de correção. Ao invés de ter que corrigir o erro ou alteração desejada em todas as linhas e células equivalentes (na situação sem função), basta alterar a definição da função para que todas as instâncias em que a função foi utilizada tenha o resultado alterado (1 correção ao invés de correções linha a linha). \
A terceira vantagem é a de usabilidade, em que a função pode ser reutilizada diversas vezes sem a necessidade de escrever separadamente aquilo que deseja ser feito toda vez. Com isso, além de evitar várias repetições de código, auxilia em mantê-lo organizado e legível.

Duas características que toda função deve ter é a palavra "def", a qual define e indica o que vai ser escrito é uma função; e o nome da função a ser utilizado, funcionando como identificador que permitirá executar o que está escrito dentro da função toda vez que esta for chamada por meio deste.

```python
# Array gerado:
np.random.seed(42)
array_normal = np.random.normal(10, 3, 5)
print("Array gerado:")
print(array_normal)
print()

# Função gerada:
def identificar_sup(obj:np.ndarray, sup:int | float) -> None:
    maiores = []
    for x in obj:
        if x > sup:
            maiores.append(float(x))
    print("Quais valores são superiores a:", sup)
    print(maiores)
    print()
    print("Quantos valores são superiores:", sup)
    print(len(maiores))
    return None

# Função executada:
identificar_sup(obj = array_normal, sup = 10)
```

## Desenvolva um programa de computador que retorne a média de um conjunto de valores.

```python
# Função gerada:
def identificar_media(obj:np.ndarray | list | tuple) -> float:
    n = 0
    media = []
    while n < len(obj):
        media.append(obj[n])
        n = n+1
    media = np.array(media)
    media = float(media.sum()/n)
    return media
```

```python
# Conferindo a função:
    # Com array
np.random.seed(42)
array_normal = np.random.normal(10, 3, 5)

print(identificar_media(obj = array_normal)) # Executando função

    # Com lista
lista_teste = [0, 10, 20, 30, 40, 50]
print(identificar_media(obj = lista_teste))

    # Com tupla
tupla_teste = (0, 20, 40, 60, 80, 100)
print(identificar_media(obj = tupla_teste))
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def divisor_simples (obj_1: int | float, obj_2: int | float) -> float:
    if type(obj_1) == int or type(obj_1) == float:
        if type(obj_2) == int or type(obj_2) == float:
            resposta = obj_1/obj_2
        else:
            print("obj_1 é inteiro ou float, mas não obj_2")
    else:
        if type(obj_2) == int or type(obj_2) == float:
            print("obj_2 é inteiro ou float, mas não obj_2")
        else:
            print("Nenhum são inteiro ou float")
    return resposta


divisor_simples(obj_1=10, obj_2=2)
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
# Definindo a função
def divisor_composto (obj_1: list | tuple, obj_2: list | tuple) -> list:
    tipos = (int, float)
    resposta = []
    for x in range(len(obj_1)):
        if type(obj_1[x]) in tipos:
            if type(obj_2[x]) in tipos:
                resposta.append(obj_1[x]/obj_2[x])
            else:
                print("Valor", x, "de obj_1 é inteiro ou float, mas não", x, "de obj_2")
        else:
            if type(obj_2[x]) in tipos:
                print("Valor", x, "de obj_2 é inteiro ou float, mas não", x, "de obj_1")
            else:
                print("Nenhum dos valores", x, "de obj_1 e obj_2 são inteiros ou float")
    return resposta
```

```python
# Testando a função com diferentes valores
divisor_composto(obj_1=[30,"30","40",80,100,"200",200,4.2],
                 obj_2=[2,4,6,"10",10,"200",200,5.2])
```
