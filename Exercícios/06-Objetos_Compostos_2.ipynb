## 1 – Um objeto composto multidimensional (matriz) pode receber  objeto composto unidimensional (vetor)?

Sim, é possível como demonstrado no código abaixo, utilizando listas e tuplas (exemplos de objetos compostos unidimensionais) para construir uma matriz.

```python
objeto_composto_unidimensional_1 = ("Carpa", "Truta", "Traíra", "Tilápia")
objeto_composto_unidimensional_2 = [7, 2, 5, 2]
objeto_composto_unidimensional_3 = [True, True, False, True]
objeto_composto_multidimensional = {
    'nome': objeto_composto_unidimensional_1,
    'massa': objeto_composto_unidimensional_2,
    'invasão': objeto_composto_unidimensional_3
}
print(objeto_composto_multidimensional)
```

## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
lista_nomes = ["SRY", "Cetn-1", "PFR2", "HER2", "GFP"]

gene0 = [len(lista_nomes[0]), [s[:2] for s in lista_nomes][0], lista_nomes[0].count('a'), 'hu' in lista_nomes[0]]
gene1 = [len(lista_nomes[1]), [s[:2] for s in lista_nomes][1], lista_nomes[1].count('a'), 'hu' in lista_nomes[1]]
gene2 = [len(lista_nomes[2]), [s[:2] for s in lista_nomes][2], lista_nomes[2].count('a'), 'hu' in lista_nomes[2]]
gene3 = [len(lista_nomes[3]), [s[:2] for s in lista_nomes][3], lista_nomes[3].count('a'), 'hu' in lista_nomes[3]]
gene4 = [len(lista_nomes[4]), [s[:2] for s in lista_nomes][4], lista_nomes[4].count('a'), 'hu' in lista_nomes[4]]

matriz_genes = [gene0, gene1, gene2, gene3, gene4]
matriz_genes
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
import pandas as pd #Importando biblioteca pandas para poder utilizar DataFrame
```

```python
novo_dic_genes = {
    lista_nomes[0]: gene0,
    lista_nomes[1]: gene1,
    lista_nomes[2]: gene2,
    lista_nomes[3]: gene3,
    lista_nomes[4]: gene4,
}
df = pd.DataFrame(novo_dic_genes)
df = df.transpose()
df.columns = [
    "Número de Caracteres",
    "Dois Primeiros Caracteres",
    "Quantidade de 'a's",
    "Presença de Caracteres 'hu'"]
df
```

```python
print(
    "Número de Caracteres =",
    df["Número de Caracteres"].sum()
)
print(
    "Quantidade de caracteres 'a' =",
    df["Quantidade de \'a\'s"].sum()
)
```

## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
df_2 = pd.read_csv('data.csv')
n_de_entradas = len(df_2)
classes_das_colunas = []
n_elementos_classe = []
for x in df_2.dtypes.value_counts().index:
    classes_das_colunas.append(x)
for x in df_2.dtypes.value_counts().values:
    n_elementos_classe.append(x)
media = [
    df_2.iloc[:,0].mean(),
    df_2.iloc[:,3].mean(),
    df_2.iloc[:,8].mean()
    ]
desvio = [
    df_2.iloc[:,0].std(),
    df_2.iloc[:,3].std(),
    df_2.iloc[:,8].std()   
]
terceiro_quartil = [
    df_2.iloc[:,0].quantile(0.75),
    df_2.iloc[:,3].quantile(0.75),
    df_2.iloc[:,8].quantile(0.75)   
]
dicionario_df = {
    "n_de_entradas": n_de_entradas,
    "classes_das_colunas": classes_das_colunas,
    "n_elementos_classe": n_elementos_classe,
    "media": media,
    "desvio": desvio,
    "terceiro_quartil": terceiro_quartil,
}
dicionario_df
```
