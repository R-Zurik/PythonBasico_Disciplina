Instancie um objeto da classe DataFrame, com os dados gerados nos dois exercícios.

```python
np.random.seed(42)
array_pacientes = np.random.randint(0,23,1000)
df_pac = pd.DataFrame(array_pacientes)
df_pac.columns = ["Nº genes diferencialmente expressos"]
df_pac.head()
```

## 03 – Categorize em grupos de 6 os dados contidos na DataFrame anterior, segundo critérios escolhidos por você. Explique os critérios.

```python
# Criando nova coluna
df_pac["Amostras"] = None

# Categorizando em 6 grupos
for x in range(len(df_pac["Nº genes diferencialmente expressos"])):
    if df_pac.iloc[x,0] <12:
        if df_pac.iloc[x,0]<4:
            df_pac.iloc[x,1] = "Amostra 1"
        elif 4<=df_pac.iloc[x,0]<8:
            df_pac.iloc[x,1] = "Amostra 2"
        else:
            df_pac.iloc[x,1] = "Amostra 3"
    else:
        if 12<=df_pac.iloc[x,0]<16:
            df_pac.iloc[x,1] = "Amostra 4"
        elif 16<=df_pac.iloc[x,0]<20:
            df_pac.iloc[x,1] = "Amostra 5"
        else:
            df_pac.iloc[x,1] = "Amostra 6"
display(df_pac.head(), df_pac.tail())
```

O critério escolhido foi dividir os grupos em 6 partes com tamanhos de intervalos iguais. Uma vez que os números inteiros do df_pac varia de 0 a 23 (24 números possíveis), dividindo em 6 partes resulta em 6 grupos de intervalo igual a 4 (0 a 3; 4 a 7; 8 a 11; 12 a 15; 16 a 19; e 20 a 23). Neste caso, não cria um peso dos intervalos e permite distribuir mais egualitariamente os valores.
Além disso, ao gerar um gráfico espera-se que, com esta divisão e forma que os valores foram gerados, seja observado barras de tamanho similares. Isto será diferente se os valores gerados tivessem sido de uma distribuição normal (maior frequência no centro e menor nas extremidades), servindo também como uma forma de comparação com os valores gerados no exercício 1.

## 04 – Defina uma função em Python que receba uma pd.Series contendo uma variável quantitativa e retorne uma DataFrame contendo a tabela de frequência.

Obtenha a tabela de frequência das variáveis criadas nos exercícios 1 e 2.

```python
# Fazendo a mesma categorização no DataFrame do exercício 1 para poder obter a tabela de frequência
    # Será utilizada o mesmo princípio de categorização, dividindo em 6 amostras de mesmo tamanho

# Primeiro, observando a informação da DataFrame para decidir os intervalos
display(df_gns.info(), df_gns["Up N Down regulation"].min(),
        df_gns["Up N Down regulation"].max())
```

Considerando que os valores mínimos e máximos equivalem a -3,24 e 3,85, fará a categorização de -3,90 à 3,90, ficando 1,3 para cada categoria: <br>
- -3,90 <= x < -2,60
- -2,60 <= x < -1,30
- -1,30 <= x < 0
- 0 <= x < 1,30
- 1,30 <= x < 2,60
- 2,60 <= x < 3,90

```python
# Segundo, a categorização em si
df_gns["Amostras"] = None

for x in range(len(df_gns["Up N Down regulation"])):
    if df_gns.iloc[x,0] < 0:
        if df_gns.iloc[x,0] < -2.60:
            df_gns.iloc[x,1] = "Amostra 1"
        elif -2.60 <= df_gns.iloc[x,0] < -1.30:
            df_gns.iloc[x,1] = "Amostra 2"
        else:
            df_gns.iloc[x,1] = "Amostra 3"
    else:
        if 0 <= df_gns.iloc[x,0] < 1.30:
            df_gns.iloc[x,1] = "Amostra 4"
        elif 1.30 <= df_gns.iloc[x,0] < 2.60:
            df_gns.iloc[x,1] = "Amostra 5"
        else:
            df_gns.iloc[x,1] = "Amostra 6"
display(df_gns.head(), df_gns.tail())
```

```python
# Definindo função que recebe uma série e retorna um DataFrame com frequência
def calcular_freq(obj:pd.Series) -> pd.DataFrame:
    df_obj = pd.DataFrame(obj.value_counts().sort_index())
    df_obj.columns=["Frequência Absoluta Simples"]
    df_obj["Frequência Relativa Simples"] = df_obj[
    "Frequência Absoluta Simples"]/df_obj["Frequência Absoluta Simples"].sum() 
    df_obj["Frequência Absoluta Acumulada"] = df_obj["Frequência Absoluta Simples"].cumsum()
    df_obj["Frequência Relativa Acumulada"] = df_obj["Frequência Relativa Simples"].cumsum() 
    return df_obj
```

```python
# Confirmando a função para o DataFrame do exercício 1
series_1 = pd.Series(df_gns["Amostras"])
df_1 = calcular_freq(obj=series_1)
df_1
```

```python
# Confirmando a função para o DataFrame do exercício 2
series_2 = pd.Series(df_pac["Amostras"])
df_2 = calcular_freq(obj=series_2)
df_2
```
