## 03 – Usando o método DataFrame.to_csv(‘nome_do_arquivo.csv’) exporte o objeto obtido no exercício 10 da aula anterior para um arquivo CSV. Usando o método pd.read_csv(‘nome_do_arquivo.csv’) instancie uma DataFrame. Use o valor ‘Unnamed: 0’ no parâmetro index_col, para ajustar o index. Arredonde os valores para até três casas decimais, usando o método round().
Recupere a DataFrame.

```python
import pandas as pd
import numpy as np
```

```python
# Recuperando o objeto criado anteriormente

# Variáveis iniciais
a = 2.0 + 3.0/5.0
b = (a+10.2)-5.0*2.0
c = a/b + 33.8
d = a/(b+c)
e = d*2

# Transformando em array
lista_1 = [a, b, c, d, e]
lista_2 = [a**0.5, b**0.5, c**0.5, d**0.5, e**0.5]
lista_3 = lista_1 + lista_2
lista_duplicada = lista_3*2
lista_array = np.array(lista_duplicada).reshape(5,4)

# Adicionando a média das colunas
media_colunas = []
x = 0
while x < 4:
    media_colunas.append(float(np.mean(lista_array[:,x]))),
    x=x+1
media_colunas
lista_array = np.append(lista_array, media_colunas)

# Transformando em DataFrame
lista_array = lista_array.reshape(-1,4)
df = round(pd.DataFrame(lista_array), 3)

# Adicionando a média das linhas
media_linhas = df.mean(axis=1)
media_linhas = round(media_linhas, 3)
df = pd.concat([df, media_linhas], axis = 1, ignore_index=True)

# Renomeando
df.rename(
    columns = {4:"Média Linhas"},
    index = {5:"Média Colunas"},
    inplace = True
)

df
```

```python
df.to_csv('nome_do_arquivo.csv')
```

## 05 – Quantos dos valores contidos na DataFrame são iguais aos valores no array?

## 09 – Quantos dos valores das médias das linhas são inferiores ou iguais à 5?

Obs 1.: Usando o método de classe, você pode converter qualquer objeto em outra classe. Ex. obj = list(array)

Obs 2.: Assim com o objeto da classe str, listas também possuem a função count()

```python
# Primeira forma possível
lista_df_2 = df_2.iloc[:,4].to_numpy().tolist()
lista_df_2
x = 0
lista_menor_igual_5 = []
while x < len(lista_df_2):
    lista_menor_igual_5.append(lista_df_2[x]<=5)
    x = x+1
lista_menor_igual_5.count(True)
```

```python
# Segunda forma possível
lista_df_2 = list(df_2.values.transpose())
(lista_df_2[4]<=5).sum().sum()
```

```python
# Terceira forma possível
lista_df_2 = list(df_2.iloc[:,4])
x = 0
lista_menor_igual_5 = []
while x < len(lista_df_2):
    lista_menor_igual_5.append(lista_df_2[x]<=5)
    x = x+1
lista_menor_igual_5.count(True)
```
