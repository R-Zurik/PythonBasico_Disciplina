## 01 – Dados as operações abaixo, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Como os operadores lógicos podem ser usados para inverter o resultado?

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0+3.0/5.0==10
q = (p+10.2)-5.0*2.0<=2.9
r = 10 <= p/q + 33.8 < 100
s = 5 != p/(q+r)
t = 33 == s*2

print("Resultado 'p and q and r and s and t':")
print(p and q and r and s and t)
print()
print("Algumas possibilidades para inverter o resultado:")
print("Substituindo o primeiro e último 'and' por 'or' =", p or q and r and s or t)
# Uma vez que 'and' ambos tem que ser True, mas o primeiro e último objeto são False
# Trocando para 'or', apenas um tem que ser True, nesse caso, o segundo e quarto objeto
print("Substituindo 'and' por 'and not' e trocando a ordem de p e q =", q and not p and r and s and not t)
# And not precisa que o primeiro seja True e o segundo False
# Funciona diretamente para s (True) e t (False), mas para p (False) e q (True) é necessário trocar a ordem
```

## 02 – Usando apenas o operador or, qual seria o resultado final das operações abaixo? Como deve ser usado o not para inverter o resultado?

<br>p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>s = ['a',1,True,'b',0,False] > ['a',True,1]
<br>t = ['a',1,True,'b',0,False] == 1
<br>u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

print("Usando apenas o operador or:")
print("p or q or r or s or t or u =", p or q or r or s or t or u)
print()
print("É necessário colocar 'or not' na relação p-q, r-s, t-u:")
print("p or not q or r or not s or t or not u =",p or not q or r or not s or t or not u)
# Para trocar True por False, 'or not' precisa ter uma relação 'False or not True'
# Dessa forma, p (False) e q (True), r (False) e s (True), t (False) e u (True)
# É necessário trocar todas essas relações e não apenas uma,
# uma vez que para "or" apenas uma relação True faz com que toda a relação seja True
```
