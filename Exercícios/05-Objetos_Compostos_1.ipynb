## 01 – Quais as vantagens da utilização de objetos compostos em detrimento dos objetos simples?

R.: Dentre as vantagens pode-se citar:
<br>1 - Velocidade mais rápida de processamento uma vez que ao instanciar a variável será utilizado uma maior região da memória RAM, em vez de regiões distribuídas para cada variável separada em objetos simples, demorando mais tempo para processar;
<br>2 - Métodos e atributos próprios de objetos compostos, que para determinadas situações podem ser preferíveis ou necessários dependendo do objetivo;
<br>3 - Possibilidade de controlar melhor as variáveis em vez de trabalhar separadamente;
<br>4 - Como cada função só permite retornar um único resultado, retornar objetos compostos permite "contornar" essa limitação apresentando mais de uma variável dentro de um objeto composto

## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Em cada uma das listas, você deve inserir o número de caracteres presentes no nome do gene, os dois primeiros caracteres do nome, quantas vezes aparece o caractere ‘a’ e se no nome aparecem os caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
gene0 = [len(lista_nomes[0]), [s[:2] for s in lista_nomes][0], lista_nomes[0].count('a'), 'hu' in lista_nomes[0]]
gene1 = [len(lista_nomes[1]), [s[:2] for s in lista_nomes][1], lista_nomes[1].count('a'), 'hu' in lista_nomes[1]]
gene2 = [len(lista_nomes[2]), [s[:2] for s in lista_nomes][2], lista_nomes[2].count('a'), 'hu' in lista_nomes[2]]
gene3 = [len(lista_nomes[3]), [s[:2] for s in lista_nomes][3], lista_nomes[3].count('a'), 'hu' in lista_nomes[3]]
gene4 = [len(lista_nomes[4]), [s[:2] for s in lista_nomes][4], lista_nomes[4].count('a'), 'hu' in lista_nomes[4]]
print(
    gene0,
    gene1,
    gene2,
    gene3,
    gene4
)
```
