p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100

```python
p = 2.0+3.0/5.0 == 10
q = (p+10.2)-5.0*2.0 <=2.9
r = 10 <= p/q + 33.8 < 100
s = 5 != p/(q+r)
t = 33 == s*2

if p and q and r and s and t:
    print("p and q and r and s and t =", True)
else:
    print("p = ", p),
    print("q = ", q),
    print("r = ", r),
    print("s = ", s),
    print("t = ", t),
    print("p and q and r and s and t =", False)
```

## Instancie um objeto da classe array numpy, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.

```python
import pandas as pd
import numpy as np
```

## 02 – Descreva o racional da sua resposta:

R.: Primeiro, como devo criar um array numpy contendo 5 valores quantitativos randômicos, é útil aplicar o método np.random.seed() para "setar" uma seed e garantir a reprodutibilidade das minhas análises. Após determinar a seed, devo então instanciar um objeto pelo método np.random.normal(), com as seguintes informações internas aos ( ): 10, 3, 5. Sendo assim, será gerado um array numpy contendo, respectivamente, média 10, desvio padrão 3, e tamanho 5. <br> Para determinar quais valores são superiores a 10, pode-se criar um objeto z vazio do tipo lista, o qual servirá para armazenar estes valores. Para armazenar pode ser utilizado um loop com "for x in array", analisando valor por valor dentro do objeto array. Neste loop, por meio de "if", pode-se avaliar se o valor x é superior a 10. ## 04 – Escreva o programa em Python:

```python
np.random.seed(42)
array_normal = np.random.normal(10, 3, 5)

quais_maior = []
for x in array_normal:
    if x > 10:
        quais_maior.append(x)

print("Array gerado:\n", array_normal, "\n\n",
      "Quais valores são superiores a 10:\n", quais_maior, "\n\n",
      "Quantos valores são superiores a 10:\n", len(quais_maior))
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

R.: Sim, seria possível. Por meio do racional proposto, bastaria trocar o valor 10 na relação if x > 10 por 20,resultando em if x > 20. Em vez de adicionar os valores superiores a 10 no objeto que seria utilizado tanto para identificar (o próprio objeto em si) quanto quantificar (este último por meio de len( )), adicionaria os valores superiores a 20. Portanto, novamente precisaria apenas retornar o próprio objeto ("Quais") e o len( ) deste ("Quantos"), como demonstrado no código abaixo:

```python
quais_maior = []
for x in array_normal:
    if x > 20:
        quais_maior.append(x)

print("Quais valores são superiores a 20:")
print(quais_maior)
print()
print("Quantos valores são superiores a 20:")
print(len(quais_maior))
```

```python
# Recuperando o objeto para exportar
lista_nomes = ["SRY", "Cetn-1", "PFR2", "HER2", "GFP"]

gene0 = [len(lista_nomes[0]), [s[:2] for s in lista_nomes][0], lista_nomes[0].count('a'), 'hu' in lista_nomes[0]]
gene1 = [len(lista_nomes[1]), [s[:2] for s in lista_nomes][1], lista_nomes[1].count('a'), 'hu' in lista_nomes[1]]
gene2 = [len(lista_nomes[2]), [s[:2] for s in lista_nomes][2], lista_nomes[2].count('a'), 'hu' in lista_nomes[2]]
gene3 = [len(lista_nomes[3]), [s[:2] for s in lista_nomes][3], lista_nomes[3].count('a'), 'hu' in lista_nomes[3]]
gene4 = [len(lista_nomes[4]), [s[:2] for s in lista_nomes][4], lista_nomes[4].count('a'), 'hu' in lista_nomes[4]]

novo_dic_genes = {
    lista_nomes[0]: gene0,
    lista_nomes[1]: gene1,
    lista_nomes[2]: gene2,
    lista_nomes[3]: gene3,
    lista_nomes[4]: gene4,
}

df = pd.DataFrame(novo_dic_genes)
df = df.transpose()
df.columns = [
    "Número de Caracteres",
    "Dois Primeiros Caracteres",
    "Quantidade de 'a's",
    "Presença de Caracteres 'hu'"]
```
